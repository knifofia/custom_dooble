## V1

```python
NB_SYMBOLS = 25
NB_CARDS = 50
NB_SYMBOL_PER_CARDS = 8

def is_allowed(card1,card2): 
    if card1 == card2:
        return True
    cpt = 0
    for symbol in card1:
        if symbol in card2:
            cpt += 1
    return cpt == 1

cards = list(combinations(range(NB_SYMBOLS), NB_SYMBOL_PER_CARDS))
for i in tqdm(range(len(cards))):
    print(i)
    if i >= len(cards):
        break
    current_card = cards[i]

    cards_to_remove = []
    for card in cards[i:]:
        if not is_allowed(current_card,card):
            cards_to_remove.append(card)
    
    print(cards_to_remove)
    for card in cards_to_remove:
        cards.remove(card)

print(len(cards))
cards
```

```python
nbSymByCard = 8
nbCards = (nbSymByCard**2) - nbSymByCard + 1
cards = []
n = nbSymByCard - 1
t = []
t.append([[(i + 1) + (j * n) for i in range(n)] for j in range(n)])
for ti in range(n - 1):
    t.append(
        [[t[0][((ti + 1) * i) % n][(j + i) % n] for i in range(n)] for j in range(n)]
    )
t.append([[t[0][i][j] for i in range(n)] for j in range(n)])
for i in range(n):
    t[0][i].append(nbCards - n)
    t[n][i].append(nbCards - n + 1)
    for ti in range(n - 1):
        t[ti + 1][i].append(nbCards - n + 1 + ti + 1)
t.append([[(i + (nbCards - n)) for i in range(nbSymByCard)]])
for ti in t:
    cards = cards + ti
print(cards)
```

# Draw cards

```python
SIZE_IMAGE = 3000
SIZE_SYMBOL = SIZE_IMAGE // 3.5
DRAW_SYMBOL_ZONE = False

CENTERS = [
    [
        int(pos[0] * SIZE_IMAGE) // 5,
        int(pos[1] * SIZE_IMAGE) // 5,
    ]
    for pos in [
        (1.7, 1.0),
        (3.2, 1.0),
        (2.5, 2.5),
        (0.9, 2.2),
        (4.1, 2.2),
        (1.2, 3.6),
        (2.5, 4.2),
        (3.8, 3.6),
    ]
]


def draw_card(symbols_ids, index) -> None:
    # Create a blank white image
    background = Image.new("RGB", (SIZE_IMAGE + 10, SIZE_IMAGE + 10), "white")
    draw_bg = ImageDraw.Draw(background)
    for center, symbol_id in zip(CENTERS, symbols_ids):
        center = center.copy()
        size_ratio = random.uniform(0.9, 1.2)

        if DRAW_SYMBOL_ZONE:
            draw_bg.ellipse(
                (
                    center[0] - (size_ratio * SIZE_SYMBOL // 2) - 2,
                    center[1] - (size_ratio * SIZE_SYMBOL // 2) - 2,
                    center[0] + (size_ratio * SIZE_SYMBOL // 2) + 2,
                    center[1] + (size_ratio * SIZE_SYMBOL // 2) + 2,
                ),
                outline="red",
            )

        symbol = Image.open(f"../img/symbols/{symbol_id}.png").convert("RGBA")

        mask = Image.new("L", symbol.size, 0)  # type: ignore
        draw_symbol = ImageDraw.Draw(mask)

        # Draw a white circle on the mask
        width, height = symbol.size
        radius = min(width, height) // 2
        draw_symbol.ellipse((0, 0, width, height), fill=255)

        # Apply the mask to the image
        circular_img = Image.new("RGBA", symbol.size)  # type: ignore
        circular_img.paste(symbol, (0, 0), mask)

        symbol = circular_img

        # symbol size

        size_symbol = int(size_ratio * SIZE_SYMBOL)
        symbol = symbol.resize(size=(size_symbol, size_symbol))  # type: ignore
        center[0] -= size_symbol // 2
        center[1] -= size_symbol // 2

        # symbol rotate
        init_image_size = symbol.size[0]
        symbol: Image = symbol.rotate(random.randint(0, 359), Image.NEAREST, expand=1)  # type: ignore
        center[0] -= (symbol.size[0] - init_image_size) // 2  # type: ignore
        center[1] -= (symbol.size[0] - init_image_size) // 2  # type: ignore

        # draw
        background.paste(symbol, (center[0], center[1]), symbol)  # type: ignore

    draw_bg.ellipse((0, 0, SIZE_IMAGE, SIZE_IMAGE), outline="black", width=2)
    background.save(f"../img/cards/card_{index}.png")


cards = pd.read_csv("symbols.csv")
for index, symbols_id in tqdm(cards.iterrows()):
    draw_card(list(symbols_id), index)
```

# cards to printable pdf

```python
pdf_filename = "../img/pages/cards.pdf"
c = canvas.Canvas(pdf_filename, pagesize=A4)
# width, height = A4
width,height = 604,850 # TODO : depends of printer

CARD_SIZE = 250
HALF_MARGIN = 2
cards_pos = [
    # center
    (width // 2 + HALF_MARGIN, height // 2 - CARD_SIZE // 2),
    (width // 2 - HALF_MARGIN - CARD_SIZE, height // 2 - CARD_SIZE // 2),
    # bottom
    (
        width // 2 + HALF_MARGIN,
        height // 2 - CARD_SIZE - CARD_SIZE // 2 - HALF_MARGIN,
    ),
    (
        width // 2 - HALF_MARGIN - CARD_SIZE,
        height // 2 - CARD_SIZE - CARD_SIZE // 2 - HALF_MARGIN,
    ),
    # top
    (
        width // 2 + HALF_MARGIN,
        height // 2 + CARD_SIZE - CARD_SIZE // 2 + HALF_MARGIN,
    ),
    (
        width // 2 - HALF_MARGIN - CARD_SIZE,
        height // 2 + CARD_SIZE - CARD_SIZE // 2 + HALF_MARGIN,
    ),
]
verso_path = "../img/verso.png"
for i in tqdm(range(10)):
    cards_path = [f"../img/cards/card_{i}.png" for i in range(i * 6, (i + 1) * 6)]
    for (x, y), card_path in zip(cards_pos, cards_path):
        if not Path(card_path).exists():
            continue
        c.drawImage(card_path, x, y, width=CARD_SIZE, height=CARD_SIZE)
    c.showPage()
    for (x, y), card_path in zip(cards_pos, cards_path):
        if not Path(card_path).exists():
            continue
        c.drawImage(verso_path, x, y, width=CARD_SIZE, height=CARD_SIZE)
    c.showPage()
c.save()
```
